```python
# Inspect dataset for:
# – numeric columns stored as text,
# – inconsistent or misspelled country names,
# – duplicate rows,
# – special characters (e.g., “–”) representing missing data.

print("Duplicate rows:", hdi_2022_df.duplicated().sum())
print("Unique countries:", hdi_2022_df['country'].unique())
print("Missing values per column:")
print(hdi_2022_df.isnull().sum())

# Define special tokens representing missing data
missing_tokens = ['–', '-', '—', '…', 'NA', 'N/A', '']

#special-character missing values
mask_special = hdi_2022_df.isin(missing_tokens)
print("Special-character missing values found per column:")
print(mask_special.sum())

# Replace special tokens with NaN
hdi_2022_df = hdi_2022_df.replace(missing_tokens, np.nan)

# Convert numeric columns stored as text
numeric_cols = ['hdi', 'gross_inc_percap', 'life_expectancy', 'gender_development']
for col in numeric_cols:
    if col in hdi_2022_df.columns:
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

# Drop rows where HDI is missing
hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])

# Columns to impute missing values with median
cols_to_impute = [
    'hdi_f', 'hdi_m', 'gross_inc_percap', 'gross_inc_percap_f', 'gross_inc_percap_m',
    'gender_development', 'gender_inequality', 'secondary_education_f_%',
    'secondary_education_m_%', 'seats_in_parliament_f_%', 'seats_in_parliament_m_%',
    'labour_participation_f_%', 'labour_participation_m_%', 'co2_emission_tons',
    'mat_footprint_percap_tons'
]

for col in cols_to_impute:
    if col in hdi_2022_df.columns:
        median_val = hdi_2022_df[col].median()
        hdi_2022_df[col] = hdi_2022_df[col].fillna(median_val)

# Strip whitespace from country names
hdi_2022_df['country'] = hdi_2022_df['country'].str.strip()

# Remove aggregate rows
aggregates = [
    'Very high human development', 'High human development', 'Medium human development',
    'Low human development', 'Arab States', 'East Asia and the Pacific',
    'Europe and Central Asia', 'Latin America and the Caribbean',
    'South Asia', 'Sub-Saharan Africa', 'World'
]
hdi_2022_df = hdi_2022_df[~hdi_2022_df['country'].isin(aggregates)]

# Fix country names
country_name_fix = {
    "Bolivia (Plurinational State of)": "Bolivia",
    "Venezuela (Bolivarian Republic of)": "Venezuela",
    "Congo (Democratic Republic of the)": "Democratic Republic of the Congo",
    "Eswatini (Kingdom of)": "Eswatini",
    "Tanzania (United Republic of)": "Tanzania",
    "Iran (Islamic Republic of)": "Iran",
    "Korea (Republic of)": "South Korea",
    "Timor-Leste": "East Timor",
    "Moldova (Republic of)": "Moldova",
    "Hong Kong, China (SAR)": "Hong Kong",
    "Micronesia (Federated States of)": "Micronesia"
}
hdi_2022_df['country'] = hdi_2022_df['country'].replace(country_name_fix)

# Final info
print(hdi_2022_df.info())
print("Missing values after cleaning:")
print(hdi_2022_df.isnull().sum())
print("Unique countries after cleaning:")
print(hdi_2022_df['country'].unique())
```

```python
# Adding HDI Category Column:
# Create a new column HDI Category
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(hdi_category)
```

```python
# 3. Outlier Detection:
outlier_df = df_south_asia[['country', 'hdi', 'gross_inc_percap']]
# Define a function to detect outliers using the 1.5 × IQR rule
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series < lower_bound) | (series > upper_bound)]
```

```python
# Detect outliers in HDI
hdi_outliers = detect_outliers(outlier_df['hdi'])
hdi_outliers
```

```python
# Detect outliers in GNI per Capita
gni_outliers = detect_outliers(outlier_df['gross_inc_percap'])
gni_outliers
```

```python
# Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.
outlier_countries = outlier_df[
    (outlier_df['hdi'].isin(hdi_outliers)) |
    (outlier_df['gross_inc_percap'].isin(gni_outliers))
]
plt.figure(figsize=(6, 5))
plt.scatter(outlier_df['gross_inc_percap'], outlier_df['hdi'], label='Normal', alpha=0.6, color= 'blue')
plt.scatter(
    outlier_countries['gross_inc_percap'],
    outlier_countries['hdi'],
    color='red',
    label='Outliers'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outliers in HDI vs GNI per Capita (South Asia)")
plt.legend()
plt.show()
```

```python
# 7. Outlier Detection:
# Detect outliers in HDI
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (df[column] < lower) | (df[column] > upper)

for df in [df_south_asia, df_middle_east]:
    df['HDI_Outlier'] = detect_outliers_iqr(df, 'hdi')
    df['GNI_Outlier'] = detect_outliers_iqr(df, 'gross_inc_percap')

# • Discuss the significance of these outliers.
```

```python
#Create scatter plots highlighting outliers in a different color.
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# South Asia
sa_normal = df_south_asia[~(df_south_asia['HDI_Outlier'] | df_south_asia['GNI_Outlier'])]
sa_outliers = df_south_asia[df_south_asia['HDI_Outlier'] | df_south_asia['GNI_Outlier']]

axes[0].scatter(
    sa_normal['gross_inc_percap'],
    sa_normal['hdi'],
    c='blue', label='Normal', alpha=0.6, s=50
)
axes[0].scatter(
    sa_outliers['gross_inc_percap'],
    sa_outliers['hdi'],
    c='red', label='Outliers', alpha=0.8, s=100, edgecolors='black', linewidth=1.5
)
axes[0].set_xlabel("GNI per Capita (PPP $)", fontsize=11)
axes[0].set_ylabel("HDI", fontsize=11)
axes[0].set_title("South Asia: HDI vs GNI (Outliers Highlighted)", fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Middle East
me_normal = df_middle_east[~(df_middle_east['HDI_Outlier'] | df_middle_east['GNI_Outlier'])]
me_outliers = df_middle_east[df_middle_east['HDI_Outlier'] | df_middle_east['GNI_Outlier']]

axes[1].scatter(
    me_normal['gross_inc_percap'],
    me_normal['hdi'],
    c='blue', label='Normal', alpha=0.6, s=50
)
axes[1].scatter(
    me_outliers['gross_inc_percap'],
    me_outliers['hdi'],
    c='red', label='Outliers', alpha=0.8, s=100, edgecolors='black', linewidth=1.5
)
axes[1].set_xlabel("GNI per Capita (PPP $)", fontsize=11)
axes[1].set_ylabel("HDI", fontsize=11)
axes[1].set_title("Middle East: HDI vs GNI (Outliers Highlighted)", fontsize=12, fontweight='bold')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle('Outlier Detection: HDI vs GNI per Capita (2020-2022)',
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
